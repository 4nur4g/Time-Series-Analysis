### QQ(Quantile-Quantile) Plot
* It is used to determine if the dataset is used to distributed in certain way.
* Usually showcases how the data fits the normal distribution.

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_copy['spx'], plot=pylab)
pylab.show()
```

* QQ plot take all the values a variable can take and arrange them in ascending order.
* Y-axis is price lowest at bottom and max at top.
* X-axis represents the theoratical quantiles of the dataset or how many standard deviations away from the mean these values are.
* the **Red Line** is what show the data points should follow if they are normally distributed.
* but in this case we can that it is not really the case since more values around the 500 value.
